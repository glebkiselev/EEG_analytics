```python
import numpy as np
import pandas as pd

data_version = 1
df = pd.read_csv(fr"E:\datasets\eeg\data{data_version}.csv")
df.head()
```

```python
df_notna = df[df["Marker value"].notna()]
df_notna
```

```python
df_length = len(df)
sequence_length = (
    df_notna.index.to_series().diff().shift(-1).fillna(0).astype(int)
)
sequence_length.iloc[-1] = df_length - sequence_length.index[-1].astype(int)
df_sequences_raw = pd.DataFrame({
    "Index start": df_notna.index.astype(int),
    "Sequence length": sequence_length.values,
    "Marker value": df_notna["Marker value"].values
})
df_sequences_raw
```

```python
df_sequences_raw["Sequence length"].mean()
```

## Сэмплинг данных

```python
batch_size = int(2048 * 1.6)  # Возьмем по 1.6 секунды (0 == вариабельные посл-сти)
start_offset = int(2048 * 0.3)  # Откинем 0.3 стартовые секунды
end_offset = int(2048 * 0.3)  # Откинем 0.3 конечные секунды
important_first = True  # Обозначим, какая часть нужнее - начало или конец, в том случае, если посл-сть меньше суммы значений сверху

df_sampled = []
for idx, row in df_sequences_raw.iterrows():
    seq_len = row['Sequence length']
    idx_start = row['Index start']
    idx_end = idx_start + seq_len
    marker_value = row['Marker value']
    
    if seq_len > batch_size > 0:
        if seq_len >= batch_size + start_offset + end_offset:
            if important_first:
                idx_start = idx_start + start_offset
                idx_end = idx_start + batch_size
            else:
                idx_end = idx_end - end_offset
                idx_start = idx_end - batch_size
        else:
            overflow = batch_size + start_offset + end_offset - seq_len
            if important_first:
                idx_start = idx_start + max(0, start_offset - overflow)
                idx_end = min(idx_start + batch_size, idx_end)
            else:
                idx_end = idx_end - max(0, end_offset - overflow)
                idx_start = max(idx_end - batch_size, idx_start)
    
    df_cur = df.iloc[idx_start:idx_end].drop(columns=[
        'Timestamp', 'Marker tag', 'Unnamed: 31',
        'Channel 25', 'Channel 26', 'Channel 27', 'Marker timestamp'
    ]).copy()
    df_cur['Sequence timestamp'] = np.arange(idx_start, idx_end, step=1, dtype='int32')
    df_cur['Marker value'] = marker_value
    df_sampled.append(df_cur)

df_sampled = pd.concat(df_sampled)
df_sampled.reset_index(drop=True, inplace=True)
df_sampled
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split

df_X = df_sampled.drop(columns=['Marker value', 'Sequence timestamp'])  # Добавить 'Sequence timestamp' для выключения временных меток в train-сете, убрать для включения
df_y = df_sampled['Marker value']

le = LabelEncoder()
df_y = le.fit_transform(df_y)

X_train, X_test, y_train, y_test = train_test_split(df_X, df_y, test_size=0.3, random_state=42)

X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
y_train = y_train.astype('int32')
y_test = y_test.astype('int32')

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
import _pickle as pkl
with open(f'scaler_v{data_version}.pkl', 'wb') as f:
    pkl.dump(sc, f)
with open(f'labels_v{data_version}.pkl', 'wb') as f:
    pkl.dump(le, f)
```

```python
from keras.layers import Dense, Input, Dropout
from keras import Sequential

model = Sequential([
    Input(shape=(X_train.shape[1],)),
    Dense(128, activation='relu'),
    Dense(len(le.classes_), activation='softmax'),
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=15)
```

```python
loss, accuracy = model.evaluate(X_test, y_test)

print(f'Loss: {loss:.4f}')
print(f'Accuracy: {accuracy:.4f}')
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

y_pred = np.argmax(model.predict(X_test), axis=-1).astype('int32')
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("Classification Report:")
print(classification_report(y_test, y_pred))
```
